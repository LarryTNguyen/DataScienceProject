# Abstract

As the COVID-19 pandemic heightened awareness of respiratory illnesses and their potential connections to environmental factors like air pollution, our group was prompted an investigation into whether poor air quality is associated with higher rates of asthma, lung cancer, and chronic obstructive pulmonary disease (COPD) incidence across U.S. states. 

In our analysis, we aimed to determine which respiratory illness, including asthma, lung cancer, or COPD, demonstrates the strongest correlation with different state’s air quality level, utilizing index (AQI) as our key metric of pollution. 

We merged and compiled state-level data on average AQI values and case counts for asthma, lung cancer, and COPD in 2020. We also conducted correlation analysis and statistical testing aiming to assess relationships between AQI and both raw respiratory disease case counts as well as population-adjusted percentages. 

As a result, our exploratory analysis indicated low correlation values between AQI and illness cases, between 0.26 to 0.44. After removing the outlier, the correlations dropped even further. In addition, the Spearman’s rank correlation tests also found no statistically significant associations between air quality and any respiratory disease type (p>0.05).

Overall, our analysis did not sufficient evidence that poor air quality, as measured by AQI, is a primary driver of differences in asthma, lung cancer, or COPD cases across U.S. states. With that being said, there must be other population and healthcare factors influencing the respiratory disease rates. Additional research which should be focused on a smaller demographic pool and geographic areas is advised to be done in order to provide further insight.

#### Description 
This dataset measures the average AQI across the U.S. by state and by CBSA from 1980 - 2021. The data tidying/cleaning involves removing rows with NaN values, filtering the average AQI by states and only including data from 2020. ### Data Tidying/Cleaning

```python
#load the dataset
aqi = pd.read_csv("AQI_2020.csv")
```

```python
#drop rows of AQI average by CBSA
aqi = aqi.drop(index = aqi.index[:33])

#only include the data from 2020 and rename columns
aqi = aqi.rename(columns = {'Years': 'State', '2020': 'avg_AQI'})
aqi = aqi[["State", "avg_AQI"]]
aqi['State'] = aqi['State'].str.replace(" (Items)", '')

#drop rows with NaN values and reset index
aqi = aqi.dropna()
aqi = aqi.reset_index()
aqi = aqi[["State", "avg_AQI"]]
```

## Asthma Cases by State
- Link: https://www.cdc.gov/asthma/archivedata/2020/2020_archived_states_territory.html#print
- Number of observations: 50
- Number of variables: 1

#### Description 
This dataset measures cases Asthma and the percent of people with Asthma in 2020. This will be useful in determining the relationship between average AQI and Asthma cases in each state. The data tidying/cleaning involves removing rows with non U.S. states, renaming columns to clearer represent the data, and dropping percentage of people with asthma.

### Data Tidying/Cleaning

```python
#load the dataset
asthma = pd.read_csv("Asthma_2020.csv")
```

```python
#remove non-state rows
non_states = ['District of Columbia', 'Guam', 'Puerto Rico']
asthma = asthma.drop(asthma[asthma['State or Territory'].isin(non_states)].index)

#reset indexing
asthma = asthma.reset_index()

#renaming columns
asthma = asthma.rename(columns = {'State or Territory': 'State', 'Number With Current Asthma': 'count_Asthma'})

#dropping percentage of 
asthma = asthma[["State", "count_Asthma"]]
```

```python
#fix variable types
asthma['count_Asthma'] = asthma['count_Asthma'].astype(str).str.replace(',', '').astype(float)
asthma.dtypes
```

```python
asthma.shape
```

```python
asthma.head(50)
```

## Lung and Brochus Cancer Cases by State
- Link: https://gis.cdc.gov/Cancer/USCS/#/AtAGlance/
- Number of observations: 48
- Number of variables: 2

#### Description 
This dataset measures cases of Lung and Brochus cancer in 2020. This will be useful in determining the relationship between average AQI and Lung and Brochus cancer cases in each state. The data tidying/cleaning involves removing rows with NaN values, non U.S. states. We will also be only keeping the columns that include states, number of cases, and population, and renaming columns to clearer represent the data.

```python
#load the dataset
lung_bronchus = pd.read_csv("Lung_Bronchus_2020.csv")
```

```python
#remove non-state rows
non_states = ['District of Columbia', 'Puerto Rico']
lung_bronchus = lung_bronchus.drop(lung_bronchus[lung_bronchus['Area'].isin(non_states)].index)

#removing rows with NaN values
lung_bronchus = lung_bronchus.drop(index = lung_bronchus.index[:2])

#reset indexing
lung_bronchus = lung_bronchus.reset_index()

#renaming columns
lung_bronchus = lung_bronchus.rename(columns = {'Area': 'State', 'Case Count': 'count_Lung_Bronchus'})

#dropping percentage of 
lung_bronchus = lung_bronchus[["State", "count_Lung_Bronchus", "Population"]]
```

```python
#fix variable types
lung_bronchus['count_Lung_Bronchus'] = lung_bronchus['count_Lung_Bronchus'].astype(int)
lung_bronchus['Population'] = lung_bronchus['Population'].astype(int)
lung_bronchus.dtypes
```

```python
lung_bronchus.shape
```

```python
lung_bronchus.head(50)
```

## COPD Cases by State
- Link: https://www.lung.org/research/trends-in-lung-disease/copd-trends-brief/data-tables/copd-prevalence-rates-by-state-gender
- Number of observations: 50
- Number of variables: 1

#### Description 
This dataset measures cases of COPD in 2020. This will be useful in determining the relationship between average AQI and COPD cases in each state. The data tidying/cleaning involves removing rows with non U.S. states, renaming columns to clearer represent the data, and only keeping columns with total case count and States.

```python
#load the dataset
copd = pd.read_csv("COPD_2020.csv")
```

```python
# removing unnecessary rows 
copd = copd.drop(index = copd.index[:1])

# #renaming column names to be more representative of the data
copd = copd.rename(columns = {'Male': 'State','Unnamed: 6': 'count_copd'})

# #removing non US states
non_states = ['District of Columbia', 'United States']
copd = copd.drop(copd[copd['State'].isin(non_states)].index)
copd = copd.reset_index()

# #keeping data of copd cases
copd = copd[["State", "count_copd"]]
copd
```

```python
#fix variable types
copd['count_copd'] = copd['count_copd'].astype(str).str.replace(',', '').astype(int)
copd.dtypes
```

```python
copd.shape
```

```python
copd.head(50)
```

## Merged Dataset

```python
#merge DataFrames based on the 'State' column
merged_data = pd.merge(aqi, asthma, on='State', how='left')
merged_data = pd.merge(merged_data, copd, on='State', how='left')
merged_data = pd.merge(merged_data, lung_bronchus, on='State', how='left')

merged_data
```

```python
merged_data.dtypes
```

```python
#creating new columns for percent of population with asthma and COPD and cancer
merged_data["perc_Lung_Bronchus"] = (merged_data["count_Lung_Bronchus"] / merged_data["Population"]) * 100
merged_data['perc_Asthma'] = (merged_data['count_Asthma'] / merged_data["Population"]) * 100
merged_data['perc_copd'] = (merged_data["count_copd"] / merged_data["Population"]) * 100
```

```python
merged_data.head()
```

### Correlation Matrices

The following correlation matrices will show the correlation coefficients between AQI and each respiratory illness. The first is based off of raw count of cases for each illness in each state and the second will be based on a percentage of cases by population in each state. The values range between -1 and 1, indicating the strength and direction of the relationship. A positive value closer to 1 indicates a strong positive correlation, while a negative value closer to -1 indicates a strong negative correlation. A value near 0 suggests a weak or no linear relationship.

### Statistical Analysis

From the previous analysis we can see that our respiratory illness data is not normally distributed, therefore the only test we can run for statistical significance is Spearmans Correlation: 

```python
#Dropping null values from any columns that have null values
df= df.dropna(subset=['count_Lung_Bronchus'])
df= df.dropna(subset=['perc_Asthma'])
df= df.dropna(subset=['perc_Lung_Bronchus'])
```

# Discusison and Conclusion
Based on the analysis, there does not appear to be a statistically significant correlation between air quality as measured by AQI and the incidence of asthma, COPD, or lung cancer when looking at raw case counts or percentages across different states. 

The exploratory analysis showed generally low correlation values between AQI and respiratory illness cases, with the highest being around 0.44 between AQI and raw asthma case counts. After removing an outlier AQI value, these correlations dropped even further. Statistical testing using Spearman's correlation found no significant correlations for any illness type with p-values all greater than 0.05.

So in summary, we did not find evidence supporting our original hypothesis that respiratory illnesses are positively correlated with poor air quality based on AQI. The analysis suggests factors other than air pollution likely play a larger role in respiratory disease rates between different states. 

This analysis was an initial look at potential connections between air quality and respiratory illness on a state-by-state basis. Going forward, analyzing the data at a more local level within states could provide additional insights. Furthermore, looking at specific pollutants rather than the general AQI index may reveal different relationships. There are likely many interplaying variables that influence respiratory health outcomes, so additional analysis controlling for relevant demographic, socioeconomic, and healthcare access factors could be worthwhile as well. But based on this study alone, air quality as measured by AQI does not appear to be a primary driver of differences in respiratory illnesses across U.S. states.

Based on the analysis so far, there are a few areas I would recommend exploring further to continue investigating the potential relationship between air quality and respiratory illness rates:

First, as mentioned above, we believe analyzing data at a more localized level is able to identify nuance trends and also minimizing confounding variables. This initial attempt of ours focused at the state-level data which can be too broad, since AQI levels most likely would not be the only factor influencing respiratory illness data. Focusing on the smaller demographics and geographic areas can help us dissect key factors behind the data easier. 

Secondly, we believe that we are able to examine more specific pollutants rather than looking only at the overall AQI, which reflects an aggregation of multiple pollutants. For example, we are able to look directly at particulate matters such as PM2.5, PM10 or even ozone levels and matching those with the respiratory illness data could possibly provide us with more insights compared to just looking at the AQI data. 

Thirdly, we should try controlling confounding variables using techniques like randomized controlled trials, matched sampling, and stratification. Many factors beyond pollution levels affect respiratory health outcomes, like access to healthcare, smoking habits, income levels etc. If we are able to identify these variables during our research, we can effectively set these variables aside and focus solely on the direct effect of air quality more precisely. 

Last but not least, we can also consider a longer period of time instead of focusing on 2020 data only. If we were able to gather enough data that delineate 5-10 years’ worth of data, we could possibly reveal longer term patterns. 

While the initial hypothesis was not supported, further refining the geographic scope, specific exposure metric, and confounding variables included could point out a clearer connection between respiratory health and air quality.
